```python
import pandas as pd
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_csv('original_rank.csv')
df
```

```python
df.fillna(0)
```

```python
df.columns
```

```python
df_rank = df[['score_risk','age','job','edu','dwell_type', 'house_type', 'marriage',
              'income_home_grp','expenditure_grp', 
              'card_count', 'loan_count','loan_amt','overdue_count','over_amt']]
df_rank
```

```python
model = sm.OLS.from_formula('score_risk ~ age+job+edu+dwell_type+house_type+marriage+\
              income_home_grp+expenditure_grp+card_count+loan_count+loan_amt+\
              overdue_count+over_amt', df_rank)
result = model.fit()
result.summary()
```
