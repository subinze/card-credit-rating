## 연체데이터에서 S98, 특수채권 변수의 이상 발견
- __overdue의 over_reason_cd 변수가 '특수채권'인 경우__
df_card와 df_loan에서 발견할 수 없음  
- 특수채권이란 ?  
연체 등의 사유로 발생한 불량채권에 대해 대손충당금을 통해 상각한 후 특수채권으로 분류해 관리함.  
일정 기간이 지나면 자산관리전문회사 등에 권리를 양도한다. = 카드사 입장에서 떼인 돈  
- __overdue의 업권코드가 S98인 경우 over_reason에 따라 loan 데이터에 있는것도 있고 없는것도 있음을 발견 (card 데이터에는 S98이 없음)__
→ 정확한 연체율 계산이 어려움

```python
import arrow
import datetime

df_card_period["card_ym"] = df_card_period["card_ym"].astype('str')

for  i in range(len(df_card_period["card_ym"])):
    date_string = df_card_period["card_ym"][i]
    date_format = '%Y%m'

    try:
        date_obj = datetime.datetime.strptime(date_string, date_format).date()
    except ValueError:
        print("Incorrect data format, should be YYYYMMDD")
    
    df_card_period["card_ym"][i] = date_obj
    
```

```python
df_card_period
```

```python
df_card_period["card_ym"][1] - df_card_period["card_ym"][2]
```

```python
from dateutil.relativedelta import relativedelta

start = datetime.datetime(2018, 12, 1, 0, 0, 0)

df_card_period["real_period"] = 0

for i in range(len(df_card_period["card_ym"])) :
    delta = relativedelta(start, df_card_period["card_ym"][i])  # 두 날짜의 차이 구하기
    result = 12 * delta.years + delta.months  # 두 날짜의 차이나는 개월수
    
    if result > 0 : 
        df_card_period["real_period"][i] = result + df_card_period["period"][i]
    else : 
        df_card_period["real_period"][i] = df_card_period["period"][i]
    
```

## 전체

1. 목표변수(:개인 연체율)에 영향을 미치는 변수 찾기  
    - EDA와 예측 모델링 기법을 활용하여 개인 연체율에 영향을 미치는 vital few 파악  
    
    
2. vital few를 활용하여 새로운 신용평가점수모델 개발  
    - vital few를 설명변수로 채택, 각 설명변수에 가중치 부여   
    
    
3. __새로운 신용점수평가모델이 실제로 성능이 좋아졌느냐에 대한 평가__
    - 각 설명변수에 임의로 가중치를 부여함  
    - 새로 산출된 신용점수에 따라, 기존 고객들 중에 신용도가 좋지 않은 고객을 추출할 수 있음  
    - 해당 고객을 기존 대출 및 연체 데이터에서 제외  
    - 나머지 데이터들을 활용하여 카드사 전체의 연체율을 계산  
    - 카드사 전체의 연체율이 낮아졌다면 더 나은 신용점수평가모델을 개발한 것으로 판단

```python
import numpy as np
import pandas as pd 
# 데이터 시각화 패키지
import matplotlib
import matplotlib.pyplot as plt
#데이터 시각화 추세 패키지
import seaborn as sns

# 그래프 주피터에서 바로 그리기 위한 설정 
%matplotlib inline 


from sklearn.tree import export_graphviz
import graphviz

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor as RFR
from sklearn.model_selection import GridSearchCV
```

```python
df_vaild = pd.read_csv("final_df_씬제거_.csv", encoding = "cp949", engine = "python" )
df_vaild_y = df_vaild["del_rate"]
df_vaild_x = df_vaild.drop("del_rate", axis = 1, inplace = False)
```

```python
df_vaild_x.columns
```

```python
df_vaild_x = df_vaild_x.drop(["chaju_id", "agegroup", "credit_class", "score_risk"], axis = 1)
```

```python
df_vaild_x_dummy = pd.get_dummies(df_vaild_x)
df_vaild_x_dummy.head()
```

```python
df_vaild_y.head()
```

```python
df_train_x, df_test_x, df_train_y, df_test_y = train_test_split(df_vaild_x_dummy, df_vaild_y, test_size = 0.3, random_state = 1234)

print("train data X size :", df_train_x.shape)
print("train data Y size :", df_train_y.shape)
print("test data X size :", df_test_x.shape)
print("test data Y size :", df_test_y.shape)

# 7:3 = 176: 76
# feature의 수 = 13
```

```python
rf_uncustomized = RFR(random_state = 1234)
rf_uncustomized.fit(df_train_x, df_train_y)

print("score on training set:", rf_uncustomized.score(df_train_x, df_train_y))
print("score on test set:", rf_uncustomized.score(df_test_x, df_test_y))
```

```python
# 설명력 결과 저장 
train_score = []
test_score = []

# n_estimators = 트리 수 변경 :1~100
para_n_tree = [n_tree*10 for n_tree in range(1,11)] # 

for v_n_estimators in para_n_tree : 
    rf = RFR(random_state = 1234, n_estimators = v_n_estimators)
    rf.fit(df_train_x, df_train_y)
    train_score.append(rf.score(df_train_x, df_train_y))
    test_score.append(rf.score(df_test_x, df_test_y))

df_score_n = pd.DataFrame()
df_score_n["parameter"] = para_n_tree
df_score_n["TrainScore"] = train_score
df_score_n["TestScore"] = test_score

df_score_n.round(3)
```

```python
plt.plot(para_n_tree, train_score, linestyle="-", label = "Train score")
plt.plot(para_n_tree, test_score, linestyle="--", label = "Test score")
```

```python
abs(df_score_n["TestScore"]-df_score_n["TrainScore"] )
```

```python
def classify(nums): 
    temp = []
    for i in nums :
        if i < 100:
            rank = "0~99점"
        elif i < 200:
            rank = "100~199점"
        elif i < 300:
            rank = "200~299점"
        elif i < 400:
            rank = "300~399점"
        elif i < 500:
            rank = "400~499점"
        elif i < 600:
            rank = "500~599점"
        elif i < 700:
            rank = "600~699점"
        elif i < 800:
            rank = "700~799점"
        elif i < 900:
            rank = "800~899점"
        elif i <=  1000:
            rank = "900~999점"
        temp.append(rank)
    return temp

rank1000 = classify(score1000)
```

## 재조정된 카드사 연체율

```python
df_loan = pd.read_csv("CreDB_loan.csv", engine = "python", encoding = "cp949")
df_overue = pd.read_csv("CreDB_overdue.csv", engine = "python", encoding = "cp949")
```

```python
# 카드사별 연체율 계산하는 함수 정의

def cal_card_overdue_rate(dfloan, dfoverdue, com_id):
    
    #연체
    df_overdue_com = dfoverdue.groupby(["chaju_id","com_id","over_amt","over_regist"]).count()
    df_overdue_company = df_overdue_com.reset_index()
    df_over_our = df_overdue_company[df_overdue_company["com_id"] == com_id]
    
    #특수채권, 주채무자 구분
    overdue_prim = df_over_our[df_over_our["over_regist"] == "주채무자(0)"]
    overdue_special = df_over_our[df_over_our["over_regist"] == "특수채권(7)"]
    print()
    
    #대출
    df_loan_com = dfloan.groupby(["chaju_id","com_id","loan_amt"]).count()
    df_loan_com.reset_index(inplace = True)
    df_loan_our = df_loan_com[df_loan_com["com_id"] == com_id]

    #우리 카드사 연체율
    df_over_rate_our = sum(df_over_our["over_amt"]) / (sum(df_loan_our["loan_amt"])+sum(overdue_special["over_amt"]))
    print("연체금 총액",sum(df_over_our['over_amt']))
    print("대출금 총액",sum(df_loan_our["loan_amt"]) +sum(overdue_special["over_amt"] ))
    print("연체율", round(df_over_rate_our*100, 2))
    
    return df_over_rate_our 
```

```python
cal_card_overdue_rate(df_loan, df_overdue, "S05-01016")
```

```python
df_lowrisk = df_nthin[df_nthin["위험점수"] <= 550]
df_lowrisk

df_temp_over = pd.merge(df_overdue, df_lowrisk, how = "right", left_on = "chaju_id", right_on = "chaju_id")
df_temp_over.rename(columns ={"over_amt_x" : "over_amt"}, inplace = True)

df_temp_loan = pd.merge(df_loan, df_lowrisk, how = "right", left_on = "chaju_id", right_on = "chaju_id")
df_temp_loan.rename(columns ={"loan_amt_x" : "loan_amt"}, inplace = True)

cal_card_overdue_rate(df_temp_loan, df_temp_over,"S05-01016")
```
