## 연체데이터에서 S98, 특수채권 변수의 이상 발견
- __overdue의 over_reason_cd 변수가 '특수채권'인 경우__
df_card와 df_loan에서 발견할 수 없음  
- 특수채권이란 ?  
연체 등의 사유로 발생한 불량채권에 대해 대손충당금을 통해 상각한 후 특수채권으로 분류해 관리함.  
일정 기간이 지나면 자산관리전문회사 등에 권리를 양도한다. = 카드사 입장에서 떼인 돈  
- __overdue의 업권코드가 S98인 경우 over_reason에 따라 loan 데이터에 있는것도 있고 없는것도 있음을 발견 (card 데이터에는 S98이 없음)__
→ 정확한 연체율 계산이 어려움

```python
import arrow
import datetime

df_card_period["card_ym"] = df_card_period["card_ym"].astype('str')

for  i in range(len(df_card_period["card_ym"])):
    date_string = df_card_period["card_ym"][i]
    date_format = '%Y%m'

    try:
        date_obj = datetime.datetime.strptime(date_string, date_format).date()
    except ValueError:
        print("Incorrect data format, should be YYYYMMDD")
    
    df_card_period["card_ym"][i] = date_obj
    
```

```python
df_card_period
```

```python
df_card_period["card_ym"][1] - df_card_period["card_ym"][2]
```

```python
from dateutil.relativedelta import relativedelta

start = datetime.datetime(2018, 12, 1, 0, 0, 0)

df_card_period["real_period"] = 0

for i in range(len(df_card_period["card_ym"])) :
    delta = relativedelta(start, df_card_period["card_ym"][i])  # 두 날짜의 차이 구하기
    result = 12 * delta.years + delta.months  # 두 날짜의 차이나는 개월수
    
    if result > 0 : 
        df_card_period["real_period"][i] = result + df_card_period["period"][i]
    else : 
        df_card_period["real_period"][i] = df_card_period["period"][i]
    
```
