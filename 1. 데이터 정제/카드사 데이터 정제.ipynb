```python
def Cal_card_company_delinquency_rate(company_code):
    #company_code = "S05-01001"
    global result
    df_raw3_drop_month=df_raw3.drop_duplicates(['chaju_id','loan_amt','loan_ym','com_id'],keep = 'first')
    
    condition = df_raw3_drop_month['com_id'] == company_code
    outdd=df_raw3_drop_month[condition]
    grouped_=outdd['loan_amt'].groupby(outdd['chaju_id']).sum()

    df_Pcard_loan_amt=pd.DataFrame(grouped_)
    df_Pcard_loan_amt.rename(columns= {'loan_amt':'Pcard_loan_amt'},inplace = True)
    df_Pcard_loan_amt

    df_raw4_drop_month=df_raw4.drop_duplicates(['chaju_id','over_amt','over_ym','com_id'],keep = 'first')
    
    condition = df_raw4_drop_month['com_id'] == company_code
    outdd=df_raw4_drop_month[condition]
    grouped_=outdd['over_amt'].groupby(outdd['chaju_id']).sum()

    df_Pcard_over_amt=pd.DataFrame(grouped_)
    df_Pcard_over_amt.rename(columns= {'over_amt':'Pcard_over_amt'},inplace = True)

    
    result = pd.merge(df_raw2,df_card_count, on = "chaju_id", how = "outer")
    result = pd.merge(result,df_card_period, on = "chaju_id", how = "outer")

    result = pd.merge(result,df_loan_period, on = "chaju_id", how = "outer")
    result = pd.merge(result,df_Pcard_loan_amt, on = "chaju_id", how = "outer")
    result = pd.merge(result,df_total_loan_amt, on = "chaju_id", how = "outer")

    result = pd.merge(result,df_over_period, on = "chaju_id", how = "outer")
    result = pd.merge(result,df_Pcard_over_amt, on = "chaju_id", how = "outer")
    result = pd.merge(result,df_total_over_amt, on = "chaju_id", how = "outer")
    result = pd.merge(result, df_overs_98case_amt, on="chaju_id", how="outer")


    result.fillna({"Pcard_loan_amt": 0, "Pcard_over_amt": 0, "loan_amt": 0, "loan_period": 0,
               "over_special_amt_Pcard": 0, "over_amt": 0, "over_period": 0, "over_normal_amt": 0,
               "over_special_amt": 0, "over_s98case_amt": 0, "연체율특채반영":0, "연체율S98반영":0}, inplace=True)
    
    result["연체율"] = result["over_amt"]/result["loan_amt"]*100
    result.fillna({"연체율" : 0},inplace = True)
    Pcard_loan_sum=result["Pcard_loan_amt"].sum()
    
    Pcard_loan_sum += result["over_special_amt_Pcard"].sum()
    
    Pcard_over_sum=result["Pcard_over_amt"].sum()

    print("{}사 대출금 총액 : \t{}".format(company_code,Pcard_loan_sum))
    print("{}사 연체금 총액 : \t{}".format(company_code,Pcard_over_sum))
    print("-"*36)
    print("{}사 연체율 : \t{:.2f}%".format(company_code, Pcard_over_sum/Pcard_loan_sum*100))
    print("="*36)
    
    
    
    
    
```
