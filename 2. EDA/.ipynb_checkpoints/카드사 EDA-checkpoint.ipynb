```python
##############################
df_raw_con_list = ["PT_THK","PT_WDTH","PT_LTH","PT_WGT","FUR_HZ_TEMP","FUR_HZ_TIME",\
                   "FUR_SZ_TEMP","FUR_SZ_TIME","FUR_EXTEMP","ROLLING_TEMP_T5","ROLLING_DESCALING"]
for column in df_raw_con_list:
    
    k=pd.get_dummies(df_raw["edu"])
    #pointbiserialr는 범주형 변수와 연속형변수의 상관성 반환
    corr,pval=stats.pointbiserialr(k,df_raw[column])
    print("{} :\t corr = {:.5} \tpval = {:.5}".format(column, corr, pval))
```

```python
from scipy import stats
k=pd.get_dummies(df_all["edu"])["대학교 졸업"]
#pointbiserialr는 범주형 변수와 연속형변수의 상관성 반환
corr,pval=stats.pointbiserialr(k,df_all["연체율S98반영"])
print("{} :\t corr = {:.5} \tpval = {:.5}".format("대학교 졸업", corr, pval))


k=pd.get_dummies(df_all["edu"])["초등학교 졸업"]
#pointbiserialr는 범주형 변수와 연속형변수의 상관성 반환
corr,pval=stats.pointbiserialr(k,df_all["연체율S98반영"])
print("{} :\t corr = {:.5} \tpval = {:.5}".format("초등학교 졸업", corr, pval))


k=pd.get_dummies(df_all["agegroup"])["40대"]
#pointbiserialr는 범주형 변수와 연속형변수의 상관성 반환
corr,pval=stats.pointbiserialr(k,df_all["연체율S98반영"])
print("{} :\t corr = {:.5} \tpval = {:.5}".format("40대", corr, pval))
```
